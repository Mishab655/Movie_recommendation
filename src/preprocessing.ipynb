```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head(1)
# movies['overview'][0]
```

```python
print(credits.shape)
print(movies.shape)
```

```python
# rename movie_id in credit.csv to join both dataframe
credits_renamed = credits.rename(columns={'movie_id':'id'})
credits_renamed
```

```python
merged_df = movies.merge(credits_renamed, on='id')
merged_df.head(3)
```

```python
merged_df.columns
```

```python
# droping unwanted columns
cleaned_df = merged_df.drop(['homepage','title_x','title_y','status','production_countries'],axis=1)
cleaned_df.head(3)
```

```python
cleaned_df.info()
```

```python
cleaned_df['overview'][0]
# to convert the text into vector, we use tfidfvector
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None,
                      strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}',
                      ngram_range=(1,3),
                      stop_words='english')
# filling none values in overview with empty string
cleaned_df['overview'] = cleaned_df['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(cleaned_df['overview'])
tfv_matrix
# it will be a sparse matrix which means having a lot of zero elments with less num of  non zero element
```

```python
tfv_matrix.shape
```

```python
# to finding the similarity between the overviews we are using sigmoid kernal, it will return a value between 0 and 1, which indicate the similarity
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
# we can use cosine matrix instead of this sigmoid kernel
```

```python
sig[0]
# here each element is the similarity of overview 0 to all overviews including overview 0
```

```python
# # reverse mapping of indices and movie title
indices = pd.Series(cleaned_df.index, index= cleaned_df['original_title']).drop_duplicates()
indices
```

```python
#  a function returning the recommended moive
def give_rec(title, sig=sig):
    index = indices[title]
    # getting the pairwise similarity scores
    sig_scores = list(enumerate(sig[index]))
    # sort the movie according to the similarity
    sig_scores = sorted(sig_scores,key=lambda x:x[1], reverse=True)
    # sort the 10 most similar movie
    sig_scores = sig_scores[1:11]
    # for movie indices
    movie_indices = [i[0] for i in sig_scores]
    return cleaned_df['original_title'].iloc[movie_indices]
    
    
```

```python
give_rec('The Dark Knight Rises')
```
